```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
file = pd.read_csv('Cleaned_2.csv',low_memory=False)
file.head(5)
```

```python
col0 = file.loc[:,'MotherAge':'MotherA'].copy()
col1 = file.loc[:,'BirthDrugAtSteroidCat'].copy()
col2 = file.loc[:,'BirthWt'].copy()
col3 = file.loc[:,'BirthGestation'].copy()
col4 = file.loc[:,'BirthMultiplicity'].copy()
col5 = file.loc[:,'BirthAdmissionTemp'].copy()
col6 = file.loc[:,'RespSpt_CPAPDur'].copy()
col7 = file.loc[:,'RespSpt_HFNCDur'].copy()
col8 = file.loc[:,'RespSpt_ConvVentDur'].copy()
col9 = file.loc[:,'RespSpt_HFOVDur'].copy()
col10 = file.loc[:,'RespSpt_NitrixOxideDur'].copy()
col11 = file.loc[:,'PostConceptAge'].copy()
col12 = file.loc[:,'DiagCVL_Duration'].copy()
col13 = file.loc[:,'OWeight'].copy()
continuous = pd.concat([col0,col1,col2,col3,col4,col5,col6,col7,col8,col9,col10,col11,col12,col13]
                        ,axis=1)
continuous.head(5)
```

```python
continuous.to_csv("Continuous.csv",index=False)
```
