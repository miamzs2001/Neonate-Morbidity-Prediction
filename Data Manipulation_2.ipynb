```python
full = pd.read_csv('2020 less than 32 weeks without hospital names (2).csv', low_memory=False)
```

```python
df2 = df.join(full['DiagIVH'])
```

```python
import numpy as np
```

```python
for ind in df2.index:
    if df2['DiagIVH'][ind]==2 or df2['DiagIVH'][ind]==4:
        df2['DiagIVHGrade'][ind]=0
df2
```

```python
df2 = df2.join(full['DiagROP'])
```

```python
for ind in df2.index:
    if df2['DiagROP'][ind]==2 or df2['DiagROP'][ind]==3:
        df2['DiagROPStage'][ind]=0
df2
```

```python
#df2 = df2.drop(['PlasticWrapAtBirth'],axis=1)
df2 = df2.join(full['PlasticWrapAtBirth'])
```

```python
df2['PlasticWrapAtBirth'] = df2['PlasticWrapAtBirth'].replace(9999,np.nan)
df2['PlasticWrapAtBirth'] = df2['PlasticWrapAtBirth'].replace(2,0)
for ind in df2.index:
    if df2['PlasticWrapAtBirth_woDry'][ind]==1:
        df2['PlasticWrapAtBirth'][ind]=2
df2
```

```python
df2 = df2.drop(['PlasticWrapAtBirth_woDry'],axis=1)
```

```python
df3 = df2.copy()
df3['DiagROPStage'] = df3['DiagROPStage'].replace(1,0)
df3['DiagROPStage'] = df3['DiagROPStage'].replace(2,1)
df3['DiagROPStage'] = df3['DiagROPStage'].replace(3,2)
df3['DiagROPStage'] = df3['DiagROPStage'].replace(4,3)
df3['DiagROPStage'] = df3['DiagROPStage'].replace(5,4)
df3['DiagROPStage'] = df3['DiagROPStage'].replace(6,5)
df3['DiagROPStage'] = df3['DiagROPStage'].replace(7,6)
df3
```

```python
df3['DiagCVL_Duration'] = df3['DiagCVL_Duration'].replace(np.nan,0)
```

```python
df3 = df3.drop(['DiagIVH','DiagROP'],axis=1)
```

```python
df3['PlasticWrapAtBirth'] = df3['PlasticWrapAtBirth'].replace(0,'No')
df3['PlasticWrapAtBirth'] = df3['PlasticWrapAtBirth'].replace(1,'Yes')
df3['PlasticWrapAtBirth'] = df3['PlasticWrapAtBirth'].replace(2,'woDry')
```

```python
df3 = pd.concat([df3,pd.get_dummies(df3['PlasticWrapAtBirth'],prefix='PlasticWrapAtBirth')],axis=1)
df3
```

```python
df3 = df3.drop(['PlasticWrapAtBirth'],axis=1)
```

```python
cleaned_v2 = df3.copy()
cleaned_v2.to_csv('Cleaned_2.csv',index=False)
```
